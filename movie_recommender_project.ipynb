```python
# Movie Recommendation System using Content-Based Filtering## 1. Import Librariesimport pandas as pdfrom sklearn.feature_extraction.text import TfidfVectorizerfrom sklearn.metrics.pairwise import linear_kernel## 2. Load Datasetmovies = pd.read_csv("movies_dataset.csv")print(movies)## 3. TF-IDF Vectorizer on genrestfidf = TfidfVectorizer(token_pattern='[a-zA-Z0-9]+')tfidf_matrix = tfidf.fit_transform(movies['genres'])## 4. Compute Similarity Matrixcosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)## 5. Recommendation Functiondef get_recommendations(title, cosine_sim=cosine_sim):    idx = movies[movies['title'] == title].index[0]    sim_scores = list(enumerate(cosine_sim[idx]))    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)    sim_scores = sim_scores[1:4]    movie_indices = [i[0] for i in sim_scores]    return movies['title'].iloc[movie_indices]## 6. Test the Recommenderprint("Recommendations for 'Inception':")print(get_recommendations('Inception'))
```
